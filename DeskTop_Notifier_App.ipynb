```python
import datetime
import requests
from win10toast import ToastNotifier
import time


def main():
    toaster = ToastNotifier()
    
    try:
        # Fetch COVID-19 data
        covidData = requests.get("https://disease.sh/v3/covid-19/countries/India")
        
        # Check if the request was successful (status code 200)
        if covidData.status_code == 200:
            # Attempt to parse the response as JSON
            try:
                data = covidData.json()
                
                # Check if 'Success' key exists in the JSON data
                if 'Success' in data:
                    data = data['Success']

                    # Show the notification with the retrieved data
                    toaster.show_toast(
                        "COVID19 Stats on {}".format(datetime.date.today()),
                        f"Total cases: {data['cases']}\nToday cases: {data['todayCases']}\n"
                        f"Today deaths: {data['todayDeaths']}\nTotal active: {data['active']}",
                        app_icon=None,  # Specify an icon path if needed
                        duration=50  # Notification stays for 50 seconds
                    )
                else:
                    print("Invalid response structure.")
            except ValueError:
                print("Error parsing JSON response.")
        else:
            print(f"Failed to fetch data. Status code: {covidData.status_code}")
    
    except requests.exceptions.RequestException as e:
        print(f"Error during request: {e}")

if __name__ == '__main__':
    # Repeat the notification every 4 hours
    while True:
        main()
        time.sleep(60 * 60 * 4)  # Sleep for 4 hours
```

```python
import datetime
import requests
from win10toast import ToastNotifier
import time

def main():
    toaster = ToastNotifier()
    
    try:
        # Fetch COVID-19 data from the new API
        print("Fetching data from the API...")
        covidData = requests.get("https://disease.sh/v3/covid-19/countries/India")
        
        # Check if the request was successful
        if covidData.status_code == 200:
            data = covidData.json()
            print("Data fetched successfully!")
            
            # Display notification with the retrieved data
            print("Displaying notification...")
            toaster.show_toast(
                "COVID19 Stats on {}".format(datetime.date.today()),
                f"Total cases: {data['cases']}\nToday cases: {data['todayCases']}\n"
                f"Today deaths: {data['todayDeaths']}\nTotal active: {data['active']}",
                app_icon=None,  # Specify icon path if needed
                duration=50  # Notification stays for 50 seconds
            )
        else:
            print(f"Failed to fetch data. Status code: {covidData.status_code}")
    
    except requests.exceptions.RequestException as e:
        print(f"Error during request: {e}")

if __name__ == '__main__':
    # Repeat the notification every 4 hours
    while True:
        main()
        time.sleep(10)  # Sleep for 4 hours
```
